# Basic Operations of NumPy & Matplotlib

## Import conventions

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Integration of NumPy & Matplotlib

```python
x = np.linspace(0, 2 * np.pi, 1000)
y = np.sin(x)
plt.plot(x, y)
plt.title('Sine from x = 0 to x = 2pi')
plt.show()
```

```python
plt.plot(y, x)
plt.show()
```
