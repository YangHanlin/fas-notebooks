```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='Linear Function')
plt.plot(x, x**2, label='Quadratic Function')
plt.plot(x, x**3, label='Cubic Function')

plt.xlabel('Value of x')
plt.ylabel('Value of y')

plt.title("Sample Plot")

plt.legend()

plt.show()
```

```python
# Demo of "Line Segment Simplication"
# So what is "Line Segment Simplication"?

y = np.random.rand(100000)
y[50000:] *= 2
y[np.logspace(1, np.log10(50000), 400).astype(int)] = -1
mpl.rcParams['path.simplify'] = True

mpl.rcParams['path.simplify_threshold'] = 0.0
plt.plot(y)
plt.show()

mpl.rcParams['path.simplify_threshold'] = 1.0
plt.plot(y)
plt.show()
```

### `matplotlib.pyplot` Module

The `matplotlib.pyplot` API mainly provides a MATLAB-style way of plotting.

> Simple as it is, the `matplotlib.pyplot` module is **not** recommended by Matplotlib official. Use the object-oriented API instead if possible.

```python
plt.plot([1, 2, 3, 4]) # with automatically-generated x values
plt.ylabel('Some Numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # 'ro' is the MATLAB-style format string; see the documentation for details
plt.axis([0, 6, 0, 20]) # [xmin, xmax, ymin, ymax]
plt.show()
```

```python
t = np.linspace(0.0, 10.0, 20)
plt.plot(t, t, 'r--', t, t**2, 'bo', t, t**3, 'g^') # plotting multiple plots in one statement
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

# For comprehension, the entries in the dictionary are printed
for key in data:
    print('data[{}] =\n{}'.format(key, data[key]))

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('Entry a')
plt.ylabel('Entry b')
plt.title('Control the hyperparameters with arrays (with dict)')
plt.show()

# Equivalent with the plot below
plt.scatter(data['a'],  data['b'], c=data['c'], s=data['d'])
plt.xlabel('Entry a')
plt.ylabel('Entry b')
plt.title('Control the hyperparameters with arrays')
plt.show()
```

```python
# Difference between title() and suptitle()

plt.suptitle('Suptitle - for Figure')
plt.subplot(121)
plt.title('Title 1 - for Axes 1')
plt.subplot(122)
plt.title('Title 2 - for Axes 2')
plt.show()
```

```python
names = ['Group A', 'Group B', 'Group C']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131) # note the special format '131' instead of '1, 3, 1'
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
# Creation of multiple figures at a time

plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title

plt.show()
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu='+str(mu)+',\ \sigma='+str(sigma)+'$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
